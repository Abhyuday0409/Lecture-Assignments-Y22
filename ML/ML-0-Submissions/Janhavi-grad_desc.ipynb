```python
## QUESTION 2 ##

import matplotlib.pyplot as plt
X = [2.3, 5, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.4, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.7, 6.8, 7.4]
Y = [18, 45, 25, 72, 30, 20, 88, 60, 81, 25, 85, 62, 41, 44, 17, 95, 30, 24, 67, 69, 30, 54, 33, 70, 85]
plt.plot(X,Y,'o') # 'o' for creating scatter plot
plt.title("Implementing Gradient Descent")
plt.xlabel("x axis")
plt.ylabel("y axis")

m=0
b=0

def grad_descent(X,Y,m,b):
    for point in zip(X,Y):
        x= point[0] # value of x at a point
        y= point[1] # value of y correspoinding to x
        
        pred_y = m*x + b # predicting value of y for given x
        error = pred_y - y
        
        # Assuming learning rate = 0.0005
        a = 0.0005
        del_m = -1 * (error*x) * a
        del_b = -1 * (error) * a
        
        #Modifying value of m and b to reduce error
        m  += del_m 
        b += del_b
    return m,b


def line(X,m,b):
    reg_x = X #List of values of x
    reg_y = [] #List of values of y
    for x in reg_x:
        y = m*x + b # calulating y prediction 
        reg_y.append(y) #appending the predicted value of y in the list
    
    # plotting the regression line for 2 sec before changing
    plt.plot(reg_x, reg_y) 
    plt.pause(1)
    
    
for i in range(1,10):
    m,b = grad_descent(X,Y,m,b)
    line(X,m,b)
    
plt.show()
```

The best solution is obtained from a tradeoff between model's ability to minimize bias and variance. Bias is defined as the difference between the predicted and correct value of of the model. On the other hand, variance shows the spread of data ("variability of model"). High biasing leads to underfitting of data (hypothesis doesn't accurately fit the dataset. High variance leads to overfitting of data (fitting is too accurate and complex). Thus, bias and variance should be kept low.


The Bias Variance Tradeoff:
If hypothesis is too simple (high bias, low variance) --> it would be more error prone
If hypothesis is too complex (high variance, low bias) --> doesn't perform well

The in-between of these two conditions is the Bias Variance Tradeoff. The algorithm can't be too complex or too simple.
Error(model) = variance(model) + bias^2(model) + variance (irreducible error)
